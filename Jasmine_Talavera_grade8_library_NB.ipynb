```python
books_ref = db.reference("books")
all_books_data = books_ref.get()

# Filter out entries where the key is a numerical string (like '0', '1', '2')
# and only keep entries where the key corresponds to the book's actual ID (like '2001', '2002', '2003').
# This assumes the 'id' field within the dictionary is the correct identifier.
# Entries like '0': {'id': 2001, ...} will be removed because '0' != '2001'.
books_data = {k: v for k, v in all_books_data.items() if isinstance(v, dict) and str(v.get('id')) == k}

print(books_data)
```

```python
for books_title, info in books_data.items():
 print(info["title"])
```

```python
for books_title, info in books_data.items():
 print("Title:", info["title"], "availability:", info["details"]["available"])
```
